```python
import pandas as pd
import requests
import time

anime = pd.read_csv("../data/anime.csv")
characters = pd.read_csv("../data/anime_characters.csv")
entities = pd.read_csv("../data/entities.csv")
```

```python
def get_anime_english_title(mal_anime_id):
    resp = requests.get(f"https://api.jikan.moe/v4/anime/{int(mal_anime_id)}")
    if resp.status_code == 429:
        time.sleep(10)
        return get_anime_english_title(mal_anime_id)
    if resp.status_code != 200:
        return None
    data = resp.json().get("data", {})
    return data.get("title_english") or data.get("title")  # fallback to japanese if no english

# Test
print(get_anime_english_title(58514))
```

```python
anime_sorted = anime.sort_values(by="popularity", ascending=True)
top2000 = anime_sorted.head(2000)

english_titles = {}
for i, anime_id in enumerate(top2000["anime_id"].unique()):
    english_title = get_anime_english_title(anime_id)
    english_titles[anime_id] = english_title
    print(f"[{i+1}] {anime_id} -> {english_title}")
    time.sleep(1.1)
    if (i + 1) % 100 == 0:
        print(f"Progress: {i+1}/{len(top2000['anime_id'].unique())}")

top2000["title_english"] = top2000["anime_id"].map(english_titles)
top2000.head(50)
```

```python
top2000.to_csv("top2000_with_english_titles.csv", index=False)
```

```python
anime = anime.merge(
    top2000[["anime_id", "title_english"]],
    on="anime_id",
    how="left"
)
anime.head()
```

```python
genres = pd.read_csv("../data/anime_genres.csv")
```

```python
genre_clean = genres[
    ~genres["genre"].str.startswith("Theme::") &
    ~genres["genre"].str.startswith("Demographic::") &
    ~genres["genre"].str.startswith("Award Winning")
]

def clean_genre(g):
    if pd.isna(g):
        return g
    

    parts = g.split(" ")
    n = len(parts)
    if n % 2 == 0 and parts[:n//2] == parts[n//2:]:
        return " ".join(parts[:n//2])
    return g

genre_clean["genre"] = genre_clean["genre"].apply(clean_genre)
genre_clean.head(50)
```

```python
genre_priority = [
    "Romance",
    "Mystery", 
    "Horror",
    "Slice of Life",
    "Adventure",
    "Action",
    "Sci-Fi",
    "Fantasy",
]

def get_primary_genre(anime_id):
    anime_genres = genre_clean[genre_clean["anime_id"] == anime_id]["genre"].tolist()
    
    # Return highest priority genre the anime has
    for genre in genre_priority:
        if genre in anime_genres:  # check against anime_genres, not genre_clean
            return genre
    
    # Fallback
    return anime_genres[0] if anime_genres else "Other"

anime["primary_genre"] = anime["anime_id"].apply(get_primary_genre)
print(anime["primary_genre"].value_counts())
```
