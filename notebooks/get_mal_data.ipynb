```python
import requests
import time
import pandas as pd
import os

def reverse_name(name):
    if pd.isna(name):
        return None
    parts = [p.strip() for p in name.split(",")]
    if len(parts) == 2:
        return f"{parts[1]} {parts[0]}"
    return name

def get_character_details(name):
    resp = requests.get(
        "https://api.jikan.moe/v4/characters",
        params={"q": name, "limit": 1}
    )
    if resp.status_code == 429:
        print("Rate limited, sleeping 60s...")
        time.sleep(60)
        return get_character_details(name)
    if resp.status_code != 200:
        return None
    data = resp.json().get("data", [])
    if not data:
        return None
    char = data[0]
    return {
        "mal_id": char["mal_id"],
        "name": char["name"],
        "biography": char.get("about")
    }

CHECKPOINT_FILE = "character_results_checkpoint.csv"

if os.path.exists(CHECKPOINT_FILE):
    checkpoint_df = pd.read_csv(CHECKPOINT_FILE)
    done_ids = set(checkpoint_df["character_id"].tolist())
    results = checkpoint_df.to_dict("records")
    print(f"Resuming from checkpoint: {len(done_ids)} already done")
else:
    done_ids = set()
    results = []

remaining = unique_chars[~unique_chars["character_id"].isin(done_ids)]
print(f"Remaining: {len(remaining)}")

for i, (_, row) in enumerate(remaining.iterrows()):
    search_name = reverse_name(row["name"])
    result = get_character_details(search_name)
    
    if result:
        result["character_id"] = row["character_id"]
    else:
        result = {"character_id": row["character_id"], "mal_id": None, "name": row["name"], "biography": None}
    
    results.append(result)
    
    # Print each result to verify
    print(f"[{i+1}] {search_name} -> mal_id: {result.get('mal_id')}")
    
    time.sleep(1.1)
    
    if (i + 1) % 500 == 0:
        pd.DataFrame(results).to_csv(CHECKPOINT_FILE, index=False)
        print(f"--- Checkpoint saved at {i+1}/{len(remaining)} ---")

pd.DataFrame(results).to_csv(CHECKPOINT_FILE, index=False)
print("Done!")
```

```python
def get_character_details_verified(name, expected_anime_title):
    resp = requests.get(
        "https://api.jikan.moe/v4/characters",
        params={"q": name, "limit": 5}  # get top 5 results instead of 1
    )
    if resp.status_code == 429:
        time.sleep(10)
        return get_character_details_verified(name, expected_anime_title)
    if resp.status_code != 200:
        return None
    
    data = resp.json().get("data", [])
    if not data:
        return None
    
    expected_lower = expected_anime_title.lower()
    
    # Try to find the result whose animeography matches the expected anime
    for char in data:
        mal_id = char["mal_id"]
        
        # Fetch full character details to check animeography
        detail_resp = requests.get(f"https://api.jikan.moe/v4/characters/{mal_id}/anime")
        if detail_resp.status_code != 200:
            continue
        
        anime_list = detail_resp.json().get("data", [])
        anime_titles = [a["anime"]["title"].lower() for a in anime_list]
        
        # Check if any anime title matches
        if any(expected_lower in title or title in expected_lower for title in anime_titles):
            return {
                "mal_id": mal_id,
                "name": char["name"],
                "biography": char.get("about")
            }
        
        time.sleep(0.4)
    
    return None  # no verified match found

# Test on Maomao
result = get_character_details_verified("Maomao", "Kusuriya no Hitorigoto")
print(result)
```

```python
RECHECK_CHECKPOINT = "recheck_checkpoint.csv"

if os.path.exists(RECHECK_CHECKPOINT):
    recheck_df = pd.read_csv(RECHECK_CHECKPOINT)
    done_ids = set(recheck_df["character_id"].tolist())
    recheck_results = recheck_df.to_dict("records")
    print(f"Resuming: {len(done_ids)} already done")
else:
    done_ids = set()
    recheck_results = []

remaining = suspicious_chars[~suspicious_chars["character_id"].isin(done_ids)]
print(f"Remaining to recheck: {len(remaining)}")

for i, (_, row) in enumerate(remaining.iterrows()):
    search_name = reverse_name(row["name"])
    result = get_character_details_verified(search_name, row["title"])
    
    if result:
        result["character_id"] = row["character_id"]
        result["verified"] = True
        print(f"[{i+1}] ✓ {row['name']} -> mal_id: {result['mal_id']}")
    else:
        # Keep original if no verified match found
        result = {
            "character_id": row["character_id"],
            "mal_id": row["mal_id"],
            "name": row["name"],
            "biography": row["biography"],
            "verified": False
        }
        print(f"[{i+1}] ✗ {row['name']} -> no verified match, keeping original")
    
    recheck_results.append(result)
    time.sleep(1.1)
    
    if (i + 1) % 100 == 0:
        pd.DataFrame(recheck_results).to_csv(RECHECK_CHECKPOINT, index=False)
        print(f"--- Checkpoint saved at {i+1}/{len(remaining)} ---")

pd.DataFrame(recheck_results).to_csv(RECHECK_CHECKPOINT, index=False)
print("Done!")
```
